## Overview

This notebook will show you how to create and query a table or DataFrame that you uploaded to DBFS. [DBFS](https://docs.databricks.com/user-guide/dbfs-databricks-file-system.html) is a Databricks File System that allows you to store data for querying inside of Databricks. This notebook assumes that you have a file already inside of DBFS that you would like to read from.

This notebook is written in **Python** so the default cell type is Python. However, you can use different languages by using the `%LANGUAGE` syntax. Python, Scala, SQL, and R are all supported.

```
# File location and type
file_location = "/FileStore/shared_uploads/a.warke001@umb.edu/bank_full.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ";"

# The applied options are for CSV files. For other file types, these will be ignored.
dataf = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(dataf)
```

```
# Create a view or table

temp_table= "bank_full_csv"

dataf.createOrReplaceTempView(temp_table)
```

```
%sql
/* Query the created temp table in a SQL cell */
select * from `bank_full_csv`
```

```
#count number of people who are married, single and divorced
dataf.groupby('marital').count().show()
```

```
%sql
/* count 0 =number of married clients who have not subscribed a term deposit */
/*  count 1 =number of married clients who have subscribed a term deposit */


SELECT count(bank_full_csv.marital) from bank_full_csv where bank_full_csv.marital=="married" group by bank_full_csv.y
```

```
%sql
/* number of married clients with different poutcome*/ 
select  bank_full_csv.poutcome, count(bank_full_csv.poutcome) from bank_full_csv where bank_full_csv.marital=="married" group by poutcome
```

```
%sql
/* number of single clients with different poutcome*/ 
select  bank_full_csv.poutcome, count(bank_full_csv.poutcome) from bank_full_csv where bank_full_csv.marital=="single" group by poutcome
```

```
%sql
/* number of divorced clients with different poutcome*/ 
select  bank_full_csv.poutcome, count(bank_full_csv.poutcome) from bank_full_csv where bank_full_csv.marital=="divorced" group by poutcome
```

```
%sql
/* number of clients with different poutcome*/ 
Select bank_full_csv.age, count(*) as subscribed from bank_full_csv where bank_full_csv.y='yes' group by bank_full_csv.age order by subscribed asc
```
